```python
import pandas as pd

# -----------------------------
# LOAD COURSES
# -----------------------------

courses_df = pd.read_csv("../data/courses.csv")

# -----------------------------
# CREATE WEEKLY ASSIGNMENTS
# -----------------------------
# 5 school weeks
# One assignment per course per week

assignment_dates = pd.to_datetime([
    "2026-02-06",
    "2026-02-13",
    "2026-02-20",
    "2026-02-27",
    "2026-03-06"
])

assignments = []

assignment_id = 1

for _, course in courses_df.iterrows():

    for week_num, due_date in enumerate(assignment_dates, start=1):

        assignments.append({
            "AssignmentID": assignment_id,
            "CourseID": course["CourseID"],
            "AssignmentName": f"Week {week_num} Assessment",
            "DueDate": due_date.date(),
            "MaxScore": 100
        })

        assignment_id += 1

# -----------------------------
# CONVERT TO DATAFRAME
# -----------------------------

assignments_df = pd.DataFrame(assignments)

# -----------------------------
# SAVE TO CSV
# -----------------------------

assignments_df.to_csv("../data/assignments.csv", index=False)
```
