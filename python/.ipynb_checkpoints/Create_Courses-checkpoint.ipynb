```python
import pandas as pd

courses = []

subjects = [
    "Mathematics",
    "English",
    "Science",
    "History",
    "Geography",
    "PE"
]

teachers = [
    "Mr Johnson",
    "Mrs Taylor",
    "Ms White",
    "Mr Brown",
    "Mrs Clark",
    "Mr Adams"
]

course_id = 1

for year_level in range(7, 13):
    for subject, teacher in zip(subjects, teachers):
        courses.append({
            "CourseID": course_id,
            "CourseName": f"Year {year_level} {subject}",
            "SubjectArea": subject,
            "YearLevel": year_level,
            "TeacherName": teacher
        })
        
        course_id += 1

courses_df = pd.DataFrame(courses)

courses_df.to_csv("../data/courses.csv", index=False)
```
