```python
import pandas as pd
import random
from faker import Faker
from datetime import date

fake = Faker("en_AU")

houses = ["Red House", "Blue House", "Green House", "Gold House"]
statuses = ["Active", "Active", "Active", "Withdrawn"]
genders = ["Male", "Female"]

# Create 100 fake students fromn years 7 to year 12

students = []

student_id = 1

for year_level in range(7, 13):  # year 7 to year 12 loop

    birth_year = 2026 - (year_level + 5)

    for _ in range(100):  # 100 student loop

        gender = random.choice(genders)

        if gender == "Male":
            first_name = fake.first_name_male()
        else:
            first_name = fake.first_name_female()

        last_name = fake.last_name()

        dob = fake.date_between(
            start_date=date(birth_year, 1, 1),
            end_date=date(birth_year, 12, 31)
        )

        student = {
            "StudentID": student_id,
            "FirstName": first_name,
            "LastName": last_name,
            "Gender": gender,
            "DateOfBirth": dob,
            "YearLevel": year_level,
            "House": random.choice(houses),
            "EnrolmentStatus": random.choice(statuses)
        }

        students.append(student)

        student_id += 1

students_df = pd.DataFrame(students) # convert to dataframe

students_df.to_csv("../data/students.csv", index=False)
```
