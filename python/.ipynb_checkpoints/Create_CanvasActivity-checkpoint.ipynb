```python
import pandas as pd
import numpy as np

# -----------------------------
# LOAD DATA
# -----------------------------

students_df = pd.read_csv("../data/students.csv")
submissions_df = pd.read_csv("../data/submissions.csv")

# -----------------------------
# CALCULATE STUDENT PERFORMANCE
# -----------------------------

student_scores = (
    submissions_df
    .groupby("StudentID")["Score"]
    .mean()
    .reset_index()
)

student_scores.rename(
    columns={"Score": "AverageScore"},
    inplace=True
)

# fill missing students with low score
student_scores["AverageScore"] = student_scores["AverageScore"].fillna(30)

# -----------------------------
# CREATE SCHOOL DATES
# -----------------------------

school_dates = pd.bdate_range(
    start="2026-02-02",
    periods=25
)

# -----------------------------
# CREATE CANVAS ACTIVITY
# -----------------------------

canvas_activity = []

activity_id = 1

np.random.seed(42)

for _, student in student_scores.iterrows():

    student_id = student["StudentID"]
    avg_score = student["AverageScore"]

    # activity multiplier based on score
    # higher score = higher activity
    activity_factor = avg_score / 70

    for activity_date in school_dates:

        # generate realistic activity
        login_count = max(
            0,
            int(np.random.normal(3 * activity_factor, 1))
        )

        page_views = max(
            0,
            int(np.random.normal(25 * activity_factor, 8))
        )

        participation_count = max(
            0,
            int(np.random.normal(4 * activity_factor, 2))
        )

        canvas_activity.append({
            "ActivityID": activity_id,
            "StudentID": student_id,
            "ActivityDate": activity_date.date(),
            "LoginCount": login_count,
            "PageViews": page_views,
            "ParticipationCount": participation_count
        })

        activity_id += 1

# -----------------------------
# CONVERT TO DATAFRAME
# -----------------------------

canvas_activity_df = pd.DataFrame(canvas_activity)

# -----------------------------
# SAVE TO CSV
# -----------------------------

canvas_activity_df.to_csv("../data/canvas_activity.csv", index=False)
```
