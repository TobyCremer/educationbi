```python
import pandas as pd
import numpy as np

# Load students
students_df = pd.read_csv("../data/students.csv")

fees = []

fee_id = 1
term = "Term 1"
amount_due = 3500
fee_date = "2026-02-02"

# Randomly choose 3% of students as outstanding
np.random.seed(42)

students_df["PaymentStatus"] = np.where(
    np.random.rand(len(students_df)) < 0.97,
    "Paid",
    "Outstanding"
)

for _, student in students_df.iterrows():

    if student["PaymentStatus"] == "Paid":
        amount_paid = amount_due
    else:
        amount_paid = 0

    fees.append({
        "FeeID": fee_id,
        "StudentID": student["StudentID"],
        "Term": term,
        "AmountDue": amount_due,
        "AmountPaid": amount_paid,
        "PaymentStatus": student["PaymentStatus"]
    })

    fee_id += 1

fees_df = pd.DataFrame(fees)

fees_df.to_csv("../data/fees.csv", index=False)
```
