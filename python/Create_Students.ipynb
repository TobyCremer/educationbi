```python
import pandas as pd
import random
import numpy as np
from faker import Faker
from datetime import date

fake = Faker("en_AU")

# -----------------------------------
# RANDOM SEED
# Makes results repeatable
# -----------------------------------

random.seed(42)
np.random.seed(42)
Faker.seed(42)

# -----------------------------------
# SETTINGS
# -----------------------------------

houses = ["Red House", "Blue House", "Green House", "Gold House"]
statuses = ["Active", "Active", "Active", "Active", "Withdrawn"]
genders = ["Male", "Female"]

# -----------------------------------
# YEAR LEVEL PROFILE
# Used later to make the dataset more interesting
# -----------------------------------

year_profiles = {
    7: {
        "profile": "Strong Transition Cohort",
        "risk_weights": [0.70, 0.20, 0.07, 0.03]
    },
    8: {
        "profile": "Stable Cohort",
        "risk_weights": [0.55, 0.30, 0.10, 0.05]
    },
    9: {
        "profile": "Disengagement Risk Cohort",
        "risk_weights": [0.30, 0.35, 0.22, 0.13]
    },
    10: {
        "profile": "Attendance Concern Cohort",
        "risk_weights": [0.35, 0.35, 0.20, 0.10]
    },
    11: {
        "profile": "Improving Senior Cohort",
        "risk_weights": [0.50, 0.32, 0.13, 0.05]
    },
    12: {
        "profile": "High Performing Senior Cohort",
        "risk_weights": [0.75, 0.18, 0.05, 0.02]
    }
}

risk_profiles = [
    "Highly Engaged",
    "Moderately Engaged",
    "At Risk",
    "Disengaged"
]

# -----------------------------------
# CREATE 600 FAKE STUDENTS
# -----------------------------------

students = []
student_id = 1

for year_level in range(7, 13):

    birth_year = 2026 - (year_level + 5)

    year_profile_name = year_profiles[year_level]["profile"]
    risk_weights = year_profiles[year_level]["risk_weights"]

    for _ in range(100):

        gender = random.choice(genders)

        if gender == "Male":
            first_name = fake.first_name_male()
        else:
            first_name = fake.first_name_female()

        last_name = fake.last_name()

        dob = fake.date_between(
            start_date=date(birth_year, 1, 1),
            end_date=date(birth_year, 12, 31)
        )

        risk_profile = np.random.choice(
            risk_profiles,
            p=risk_weights
        )

        student = {
            "StudentID": student_id,
            "FirstName": first_name,
            "LastName": last_name,
            "Gender": gender,
            "DateOfBirth": dob,
            "YearLevel": year_level,
            "House": random.choice(houses),
            "EnrolmentStatus": random.choice(statuses),
            "RiskProfile": risk_profile,
            "YearProfile": year_profile_name
        }

        students.append(student)
        student_id += 1

# -----------------------------------
# CONVERT TO DATAFRAME
# -----------------------------------

students_df = pd.DataFrame(students)

# -----------------------------------
# SAVE CSV
# -----------------------------------

students_df.to_csv("../data/students.csv", index=False)

print("\nSaved to ../data/students.csv")
```
