```python
import pandas as pd

# Load CSVs
students_df = pd.read_csv("../data/students.csv")
courses_df = pd.read_csv("../data/courses.csv")

enrolments = []
enrolment_id = 1

for _, student in students_df.iterrows():

    student_id = student["StudentID"]
    year_level = student["YearLevel"]

    matching_courses = courses_df[courses_df["YearLevel"] == year_level]

    for _, course in matching_courses.iterrows():
        enrolments.append({
            "EnrolmentID": enrolment_id,
            "StudentID": student_id,
            "CourseID": course["CourseID"],
            "EnrolmentDate": "2026-02-02"
        })

        enrolment_id += 1

enrolments_df = pd.DataFrame(enrolments)

enrolments_df.to_csv("../data/enrolments.csv", index=False)
```
