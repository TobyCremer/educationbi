```python
import pandas as pd
import numpy as np

# -----------------------------
# LOAD STUDENTS
# -----------------------------

students_df = pd.read_csv("../data/students.csv")

# -----------------------------
# CREATE SCHOOL DATES
# -----------------------------
# 5 school weeks from Monday 02/02/2026
# B = business days only, Monday to Friday

school_dates = pd.bdate_range(
    start="2026-02-02",
    periods=25
)

# -----------------------------
# ATTENDANCE SETTINGS
# -----------------------------

attendance_statuses = ["Present", "Late", "Absent"]
attendance_probabilities = [0.95, 0.03, 0.02]

attendance = []
attendance_id = 1

np.random.seed(42)

# -----------------------------
# CREATE ATTENDANCE RECORDS
# -----------------------------

for attendance_date in school_dates:

    for _, student in students_df.iterrows():

        status = np.random.choice(
            attendance_statuses,
            p=attendance_probabilities
        )

        if status == "Present":
            periods_missed = 0
        elif status == "Late":
            periods_missed = 1
        else:
            periods_missed = 6

        attendance.append({
            "AttendanceID": attendance_id,
            "StudentID": student["StudentID"],
            "AttendanceDate": attendance_date.date(),
            "Status": status,
            "PeriodsMissed": periods_missed
        })

        attendance_id += 1

# -----------------------------
# CONVERT TO DATAFRAME
# -----------------------------

attendance_df = pd.DataFrame(attendance)

# -----------------------------
# SAVE TO CSV
# -----------------------------

attendance_df.to_csv("../data/attendance.csv", index=False)
```
